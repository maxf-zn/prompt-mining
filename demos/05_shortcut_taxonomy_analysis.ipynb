# Shortcut Taxonomy Analysis (Paper Reproduction)

This notebook demonstrates how to reproduce the **shortcut taxonomy** (Table 2 / Section 4.3) from the LODO paper:

> *"30% of top SAE features are dataset-dependent shortcuts"*

## Overview

The shortcut taxonomy classifies classifier features into 4 quadrants (Q1-Q4) based on two dimensions:

| | **Dataset-Predictive** | **Not Dataset-Predictive** |
|---|---|---|
| **LODO Shortcut** | Q1: Pure Shortcuts | Q2: Context-Dependent |
| **LODO Stable** | Q3: Robust but Correlated | Q4: Truly Generalizable |

**Key insight**: Q2 (context-dependent shortcuts) would be missed by simply checking dataset-predictiveness alone.

### Methodology

1. **LODO Coefficient Retention**: Train classifiers holding out each dataset; measure how much each feature's coefficient changes
2. **Firing Ratio**: Compute `mal_firing_rate / ben_firing_rate` - low ratio (<1.5x) indicates weak class separation (dataset-predictive)
3. **Taxonomy Assignment**: Cross-reference retention with firing ratio

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import LeaveOneGroupOut

from prompt_mining.classifiers import (
    ClassificationDataset,
    LinearClassifier,
    LinearConfig,
)
```

## 1. Load SAE Features

Load pre-extracted SAE features from the classification dataset. The SAE features are sparse (131K dimensions) and interpretable.

```python
# Load dataset - update path to your data
DATA_PATH = "/path/to/activations"
dataset = ClassificationDataset.from_path(DATA_PATH)

# Load SAE features (sparse for memory efficiency)
data = dataset.load(layer=27, space='sae', position=[-5], return_sparse=True)
print(dataset.summary(data))

# Identify pure-class datasets for analysis
dataset_class_rates = {}
for ds in np.unique(data.datasets):
    ds_mask = data.datasets == ds
    mal_rate = data.y[ds_mask].mean()
    dataset_class_rates[ds] = mal_rate

pure_mal_datasets = {ds for ds, rate in dataset_class_rates.items() if rate > 0.95}
pure_ben_datasets = {ds for ds, rate in dataset_class_rates.items() if rate < 0.05}
print(f"\nPure malicious datasets: {pure_mal_datasets}")
print(f"Pure benign datasets: {pure_ben_datasets}")
```

## 2. Train Baseline Class Classifier

Train a classifier to predict malicious vs benign. We use `LinearClassifier` wrapper which:
- Applies L2 normalization (important for SAE features)
- Uses `class_weight='balanced'` by default
- Handles sparse matrices correctly

```python
# Train baseline class classifier using the LinearClassifier wrapper
# This matches the configuration from the paper analysis
class_clf = LinearClassifier(LinearConfig(
    model='logistic',
    normalize='l2',  # L2 normalization for SAE features
    C=1.0,
))
class_clf.fit(data.X, data.y)
class_coefs = class_clf.coef_[0]

print(f"Class classifier trained")
print(f"Coefficient shape: {class_coefs.shape}")

# Get top-K features by positive coefficient (predict malicious)
TOP_K = 50
class_top_pos = set(np.argsort(class_coefs)[-TOP_K:])
print(f"\nTop {TOP_K} positive (malicious-predictive) features identified")
```

## 3. Compute Firing Rate Analysis

For each top feature, compute the **firing ratio** = `mal_firing_rate / ben_firing_rate`.

- **Low ratio (<1.5x)**: Feature fires similarly on both classes → weak class separation → likely relies on dataset identity
- **High ratio (≥1.5x)**: Feature fires much more on malicious → genuinely class-discriminative

```python
def analyze_feature_firing(feat_idx, X, y, datasets):
    """Compute firing statistics for a single feature."""
    feat_vals = X[:, feat_idx].toarray().flatten()
    
    mal_mask = y == 1
    ben_mask = y == 0
    
    # Firing rate = fraction of samples where feature > 0
    mal_firing = (feat_vals[mal_mask] > 0).mean()
    ben_firing = (feat_vals[ben_mask] > 0).mean()
    firing_ratio = mal_firing / max(ben_firing, 1e-6)
    
    # Per-dataset firing rates (for top dataset identification)
    ds_firing = {}
    for ds in np.unique(datasets):
        ds_mask = datasets == ds
        ds_firing[ds] = (feat_vals[ds_mask] > 0).mean()
    
    top_ds = sorted(ds_firing.items(), key=lambda x: -x[1])[0]
    
    return {
        'idx': feat_idx,
        'mal_firing': mal_firing,
        'ben_firing': ben_firing,
        'firing_ratio': firing_ratio,
        'top_dataset': top_ds[0],
        'top_ds_firing': top_ds[1],
    }

# Analyze all top features
firing_analysis = []
for feat_idx in class_top_pos:
    result = analyze_feature_firing(feat_idx, data.X, data.y, data.datasets)
    result['coefficient'] = class_coefs[feat_idx]
    firing_analysis.append(result)

firing_df = pd.DataFrame(firing_analysis)
firing_df = firing_df.sort_values('coefficient', ascending=False)

print("=== TOP 20 POSITIVE FEATURES (Predict Malicious) ===")
print(firing_df[['idx', 'coefficient', 'firing_ratio', 'mal_firing', 'ben_firing', 'top_dataset']].head(20).to_string(index=False))
```

```python
# Identify firing-rate based shortcuts:
# Features with LOW firing ratio (<1.5x) but HIGH positive coefficient
# These likely depend on dataset-class confounding
FIRING_RATIO_THRESHOLD = 1.5

firing_rate_shortcuts = firing_df[firing_df['firing_ratio'] < FIRING_RATIO_THRESHOLD]
print(f"=== FIRING RATE SHORTCUTS: Low firing ratio (<{FIRING_RATIO_THRESHOLD}x) ===")
print(f"Found {len(firing_rate_shortcuts)} / {len(firing_df)} features ({len(firing_rate_shortcuts)/len(firing_df)*100:.0f}%)\n")

for _, row in firing_rate_shortcuts.head(10).iterrows():
    top_ds = row['top_dataset']
    ds_class = 'PURE_MAL' if top_ds in pure_mal_datasets else ('PURE_BEN' if top_ds in pure_ben_datasets else 'MIXED')
    print(f"Feature {row['idx']:6.0f}: coef={row['coefficient']:6.2f}, ratio={row['firing_ratio']:.2f}x, "
          f"top_ds={top_ds} ({ds_class})")
```

## 4. LODO Coefficient Retention

For each dataset, train a classifier **without** that dataset and compare coefficients to baseline.

**Retention** = `min(fold_coef / baseline_coef)` across all folds. Low retention (<50%) means the feature depends on specific datasets.

```python
# Train LODO classifiers using LeaveOneGroupOut
logo = LeaveOneGroupOut()

# Store LODO coefficients for each top feature
lodo_coefs = {feat: {'baseline': class_coefs[feat], 'holdout': {}} for feat in class_top_pos}

n_datasets = len(np.unique(data.datasets))
print(f"Training LODO classifiers across {n_datasets} datasets...")
for train_idx, test_idx in logo.split(data.X, data.y, data.datasets):
    held_out_ds = data.datasets[test_idx[0]]
    
    # Train classifier with same config as baseline
    clf_lodo = LinearClassifier(LinearConfig(model='logistic', normalize='l2', C=1.0))
    clf_lodo.fit(data.X[train_idx], data.y[train_idx])
    
    # Store coefficients for top features
    for feat in class_top_pos:
        lodo_coefs[feat]['holdout'][held_out_ds] = clf_lodo.coef_[0, feat]
    
    print(f"  Held out: {held_out_ds}")

print(f"\nTrained {n_datasets} LODO classifiers")
```

```python
# Compute LODO retention for each top feature
lodo_retention = {}
lodo_dependent_ds = {}

for feat in class_top_pos:
    baseline = lodo_coefs[feat]['baseline']
    holdout_coefs = lodo_coefs[feat]['holdout']
    
    # Find minimum coefficient across all holdouts
    min_coef = min(holdout_coefs.values())
    min_ds = min(holdout_coefs, key=holdout_coefs.get)
    
    # Retention = min_coef / baseline (for positive baseline coefficients)
    if baseline > 0:
        retention = min_coef / baseline
    else:
        retention = float('inf')
    
    lodo_retention[feat] = retention
    lodo_dependent_ds[feat] = min_ds

# Categorize by retention threshold
RETENTION_THRESHOLD = 0.50

lodo_shortcuts = set(f for f in class_top_pos if lodo_retention[f] < RETENTION_THRESHOLD)
lodo_stable = set(f for f in class_top_pos if lodo_retention[f] >= RETENTION_THRESHOLD)

print("=== LODO Retention Categories ===")
print(f"LODO shortcuts (retention < {RETENTION_THRESHOLD:.0%}): {len(lodo_shortcuts)}")
print(f"LODO stable (retention >= {RETENTION_THRESHOLD:.0%}): {len(lodo_stable)}")
```

## 5. Build the Shortcut Taxonomy (Q1-Q4)

Now we combine **LODO retention** with **firing ratio** to create the 2x2 taxonomy:

- **Q1 (Pure Shortcuts)**: Low retention AND low firing ratio (dataset-predictive)
- **Q2 (Context-Dependent)**: Low retention BUT high firing ratio (not dataset-predictive)  
- **Q3 (Robust but Correlated)**: High retention BUT low firing ratio
- **Q4 (Truly Generalizable)**: High retention AND high firing ratio

```python
# Build firing ratio lookup
firing_ratio_lookup = dict(zip(firing_df['idx'].astype(int), firing_df['firing_ratio']))

# Dataset-predictive = low firing ratio (weak class separation)
dataset_predictive = set(f for f in class_top_pos if firing_ratio_lookup.get(f, 0) < FIRING_RATIO_THRESHOLD)
not_dataset_predictive = class_top_pos - dataset_predictive

# Build 2x2 taxonomy
# Q1: LODO shortcut AND dataset-predictive (low firing ratio)
q1 = lodo_shortcuts & dataset_predictive
# Q2: LODO shortcut but NOT dataset-predictive (high firing ratio)
q2 = lodo_shortcuts - dataset_predictive
# Q3: LODO stable but IS dataset-predictive
q3 = lodo_stable & dataset_predictive
# Q4: LODO stable AND NOT dataset-predictive (truly generalizable)
q4 = lodo_stable - dataset_predictive

print("=" * 70)
print("SHORTCUT TAXONOMY (Q1-Q4)")
print("=" * 70)
print(f"\nRetention threshold: {RETENTION_THRESHOLD:.0%}")
print(f"Firing ratio threshold: {FIRING_RATIO_THRESHOLD}x\n")

print(f"{'':25s} | {'Dataset-Predictive':^20s} | {'NOT Predictive':^20s}")
print(f"{'':25s} | {'(firing ratio<1.5x)':^20s} | {'(firing ratio>=1.5x)':^20s}")
print("-" * 70)
print(f"{'LODO Shortcut (<50%)':25s} | {'Q1: ' + str(len(q1)):^20s} | {'Q2: ' + str(len(q2)):^20s}")
print(f"{'LODO Stable (>=50%)':25s} | {'Q3: ' + str(len(q3)):^20s} | {'Q4: ' + str(len(q4)):^20s}")
print("-" * 70)

total_shortcuts = len(lodo_shortcuts)
print(f"\nTotal LODO shortcuts (Q1+Q2): {total_shortcuts} ({total_shortcuts/TOP_K*100:.0f}%)")
print(f"Total dataset-predictive (Q1+Q3): {len(dataset_predictive)}")

# Key insight from paper
if total_shortcuts > 0:
    pct_context_dependent = len(q2) / total_shortcuts * 100
    print(f"\n** {pct_context_dependent:.0f}% of shortcuts (Q2) would be missed by firing ratio alone **")
```

## 6. Detailed Feature Analysis

```python
# Show features in each quadrant
quadrants = [
    ('Q1: Pure Shortcuts', q1),
    ('Q2: Context-Dependent', q2),
    ('Q3: Robust+Correlated', q3),
    ('Q4: Truly Generalizable', q4),
]

for name, features in quadrants:
    print(f"\n{'='*60}")
    print(f"{name} ({len(features)} features)")
    print(f"{'='*60}")
    
    # Sort by coefficient magnitude
    sorted_feats = sorted(features, key=lambda f: abs(class_coefs[f]), reverse=True)
    
    for feat in sorted_feats[:5]:
        ret = lodo_retention[feat]
        dep_ds = lodo_dependent_ds[feat]
        fr = firing_ratio_lookup.get(feat, 0)
        ds_type = 'MAL' if dep_ds in pure_mal_datasets else ('BEN' if dep_ds in pure_ben_datasets else 'MIX')
        print(f"  Feature {feat:6d}: coef={class_coefs[feat]:+.2f}, "
              f"retention={ret:.0%}, firing_ratio={fr:.2f}x, depends='{dep_ds}' ({ds_type})")
```

## 7. Visualization

```python
# Build dataframe for plotting
plot_data = []
for feat in class_top_pos:
    fr = firing_ratio_lookup.get(feat, 0)
    is_dataset_predictive = fr < FIRING_RATIO_THRESHOLD
    is_shortcut = feat in lodo_shortcuts
    
    if is_shortcut and is_dataset_predictive:
        quadrant = 'Q1: Pure Shortcuts'
    elif is_shortcut and not is_dataset_predictive:
        quadrant = 'Q2: Context-Dependent'
    elif not is_shortcut and is_dataset_predictive:
        quadrant = 'Q3: Robust+Correlated'
    else:
        quadrant = 'Q4: Truly Generalizable'
    
    plot_data.append({
        'feature_idx': feat,
        'coefficient': class_coefs[feat],
        'retention': lodo_retention[feat],
        'firing_ratio': fr,
        'quadrant': quadrant,
    })

df = pd.DataFrame(plot_data)
```

```python
# Scatter plot: Firing Ratio vs LODO Retention (matches paper Figure)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

colors = {
    'Q1: Pure Shortcuts': 'red',
    'Q2: Context-Dependent': 'orange', 
    'Q3: Robust+Correlated': 'blue',
    'Q4: Truly Generalizable': 'green'
}

# Plot 1: Firing Ratio vs Retention
ax1 = axes[0]
for quadrant, color in colors.items():
    subset = df[df['quadrant'] == quadrant]
    ax1.scatter(subset['firing_ratio'], subset['retention'], 
                c=color, label=quadrant, s=subset['coefficient']*8, alpha=0.7, edgecolors='black')

ax1.axhline(y=RETENTION_THRESHOLD, color='gray', linestyle='--', 
            label=f'LODO threshold ({RETENTION_THRESHOLD:.0%})')
ax1.axvline(x=FIRING_RATIO_THRESHOLD, color='gray', linestyle=':', 
            label=f'Firing ratio threshold ({FIRING_RATIO_THRESHOLD}x)')
ax1.axhline(y=0, color='red', linestyle='-', alpha=0.3, label='Sign flip')
ax1.set_xlabel('Firing Ratio (mal/ben)')
ax1.set_ylabel('LODO Retention')
ax1.set_title('Shortcut Taxonomy: Firing Ratio vs LODO Retention')
ax1.legend(loc='lower right', fontsize=8)
ax1.set_ylim(-0.5, 1.5)

# Plot 2: Taxonomy pie chart
ax2 = axes[1]
counts = [len(q1), len(q2), len(q3), len(q4)]
labels = [f'Q1: Pure Shortcuts\n({len(q1)})', f'Q2: Context-Dependent\n({len(q2)})',
          f'Q3: Robust+Correlated\n({len(q3)})', f'Q4: Truly Generalizable\n({len(q4)})']
pie_colors = ['red', 'orange', 'blue', 'green']

wedges, texts, autotexts = ax2.pie(counts, labels=labels, colors=pie_colors, 
                                   autopct='%1.0f%%', startangle=90)
ax2.set_title(f'Shortcut Taxonomy (Top-{TOP_K} Features)')

plt.tight_layout()
plt.show()
```

```python
# Retention distribution histogram
fig, ax = plt.subplots(figsize=(10, 5))

shortcuts_ret = df[df['retention'] < RETENTION_THRESHOLD]['retention']
stable_ret = df[df['retention'] >= RETENTION_THRESHOLD]['retention']

ax.hist(shortcuts_ret, bins=20, alpha=0.7, label=f'Shortcuts ({len(shortcuts_ret)})', color='red')
ax.hist(stable_ret, bins=20, alpha=0.7, label=f'Stable ({len(stable_ret)})', color='green')
ax.axvline(x=RETENTION_THRESHOLD, color='black', linestyle='--', linewidth=2, 
           label=f'Threshold ({RETENTION_THRESHOLD:.0%})')

ax.set_xlabel('LODO Coefficient Retention')
ax.set_ylabel('Number of Features')
ax.set_title(f'Distribution of LODO Retention (Top-{TOP_K} Features)')
ax.legend()

plt.tight_layout()
plt.show()
```

## 8. Which Datasets Create Shortcuts?

Analyze which datasets are most responsible for shortcut features.

```python
# Count which datasets shortcuts depend on
dependency_counts = {}
for feat in lodo_shortcuts:
    ds = lodo_dependent_ds[feat]
    dependency_counts[ds] = dependency_counts.get(ds, 0) + 1

print("Datasets that shortcut features depend on:")
print("=" * 60)
for ds, count in sorted(dependency_counts.items(), key=lambda x: -x[1]):
    # Get dataset class composition
    ds_mask = data.datasets == ds
    mal_rate = data.y[ds_mask].mean()
    ds_type = "PURE_MAL" if mal_rate > 0.95 else "PURE_BEN" if mal_rate < 0.05 else "MIXED"
    print(f"  {ds:25s}: {count:2d} features ({ds_type}, {mal_rate:.0%} malicious)")

print(f"\n** Pure-class datasets (100% mal/benign) create the most shortcuts **")
```
